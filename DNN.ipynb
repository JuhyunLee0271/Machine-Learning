```python
from tensorflow import keras
(train_input, train_target), (test_input, test_target) =\
    keras.datasets.fashion_mnist.load_data()
```

```python
from sklearn.model_selection import train_test_split

# 데이터 정규화 후 train, validation set으로 나눔
train_scaled = train_input / 255.0
train_scaled = train_input.reshape(-1, 28*28)
train_scaled, val_scaled, train_target, val_target = train_test_split (
    train_scaled, train_target, test_size = 0.2, random_state = 42
)
```

```python
# dense1 : 은닉층, 첫 번째 layer이므로 입력의 크기를 지정해야 함, 활성화 함수로 sigmoid 사용
# dense2 : 출력층, 활성화 함수로 softmax 사용
# 심층 신경망 (Deep Neural Network)

model = keras.Sequential()
model.add(keras.layers.Dense(500, activation='sigmoid', input_shape=(784,),name='hidden1'))
model.add(keras.layers.Dense(100, activation='sigmoid', name='hidden2'))
model.add(keras.layers.Dense(10, activation='softmax',name='output'))
model.summary()

# model의 compile 설정
model.compile(loss='sparse_categorical_crossentropy',metrics='accuracy')
model.fit(train_scaled, train_target, epochs=5)
```

```python
model.evaluate(val_scaled, val_target)
```

```python
# hidden layer에 Relu activation 함수 사용
# Flatten layer: 입력층에 대해 1차원으로 펼쳐줌, 입력 데이터에 대해 reshape(-1)을 해주지 않아도 됨 
(train_input, train_target), (test_input, test_target) =\
    keras.datasets.fashion_mnist.load_data()

train_scaled = train_input / 255.0
train_scaled, val_scaled, train_target, val_target = train_test_split (
    train_scaled, train_target, test_size=0.2, random_state=42
)


model = keras.Sequential()
model.add(keras.layers.Flatten(input_shape=(28,28)))
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))

model.summary()
# model.compile(loss='sparse_categorical_crossentropy',metrics='accuracy')
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics='accuracy')
model.fit(train_scaled, train_target, epochs=5)
```

```python
model.evaluate(val_scaled, val_target)
```
